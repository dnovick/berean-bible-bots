## 1. Overview & Key Statistics

The **Hithpael** is the reflexive/reciprocal intensive stem of Biblical Hebrew.
Its diagnostic marker is the **הִתְ- prefix** before R1 (with metathesis/assimilation
rules when R1 is a sibilant: הִשְׁתַּמֵּר ← הִתְ + שָׁמַר).

**Five main semantic functions:**
- **Reflexive of Piel:** הִתְקַדֵּשׁ 'consecrated oneself' (← Piel קִדֵּשׁ)
- **Reciprocal:** הִתְרָאוּ 'saw one another'
- **Iterative/Frequentative:** הִתְהַלֵּךְ 'walked about / walked continually'
- **Tolerative/Reflexive:** הִתְמַכֵּר 'sold oneself'
- **Denominative:** הִתְנַבֵּא 'acted as a prophet'

Notable Hithpael roots: הִתְפַּלֵּל (pray, lit. intercede for oneself),
הִתְהַלֵּךְ (walk about/live), הִתְקַדֵּשׁ (consecrate oneself).
